# Data Cleaning

## Fill empty data

```python
def fill_na(x: pd.DataFrame):
  data_nan = x[x.isna().any(axis=1)]
  data_nan_index = data_nan.index
  data_nan.loc[:,"price_original"] = data_nan["price_original"].fillna("Rp" + data_nan["price_current"], inplace=False)
  data_nan.loc[:,"discount"] = data_nan["discount"].fillna("-0%", inplace=False)

  x.loc[data_nan_index] = data_nan
  return x
```

```python
electronics_data_filled = electronics_data.copy()
beauty_data_filled = beauty_data.copy()

electronics_data_filled = fill_na(electronics_data_filled)
beauty_data_filled = fill_na(beauty_data_filled)
```

## Convert data type

```python
electronics_data_converted = electronics_data_filled.copy()
beauty_data_converted = beauty_data_filled.copy()

# "rating" column
electronics_data_converted["rating"] = electronics_data_filled["rating"].astype(float)
beauty_data_converted["rating"] = beauty_data_filled["rating"].astype(float)

# "sold_quantity" column
def convert_sold(val):
  return int(
      val.lower()
        .replace("sold", "")
        .replace(" ", "")
        .replace("k", "000")
        .replace(".", "")
        .replace("m", "000000")
  )

electronics_data_converted["sold_quantity"] = electronics_data_converted["sold_quantity"].apply(convert_sold)
beauty_data_converted["sold_quantity"] = beauty_data_converted["sold_quantity"].apply(convert_sold)

# "price_current" column
electronics_data_converted["price_current"] = electronics_data_converted["price_current"].str.replace(".", "").astype(int)
beauty_data_converted["price_current"] = beauty_data_converted["price_current"].str.replace(".", "").astype(int)

# "price_original" column
def convert_original(val):
  return int(
      val.lower()
        .replace("rp", "")
        .replace(".", "")
  )
electronics_data_converted["price_original"] = electronics_data_converted["price_original"].apply(convert_original)
beauty_data_converted["price_original"] = beauty_data_converted["price_original"].apply(convert_original)

# "discount" column
def convert_discount(val):
  return int(
      val.lower()
        .replace("%", "")
        .replace("-", "")
  )

electronics_data_converted["discount"] = electronics_data_converted["discount"].apply(convert_discount)
beauty_data_converted["discount"] = beauty_data_converted["discount"].apply(convert_discount)
```

```python
import re
import nltk
from nltk import word_tokenize
nltk.download("punkt_tab")

# Text Extraction Function
def text_extraction(data: pd.DataFrame, pattern_norm: dict = None, use_normalization: bool = False) -> pd.DataFrame:
  if use_normalization and pattern_norm is None:
    raise Exception("Diperlukan pattern untuk normalisasi.")

  data_cp = data.copy()

  # Character count
  data_cp["char_punc_count"] = data_cp["name"].str.len()

  # Alphabet count
  data_cp["alpha_count"] = data_cp["name"].str.count(r"[a-zA-Z]")

  # Punctuation count
  data_cp["punc_count"] = data_cp["name"].str.count(r"[^\w\s]")

  # Jumlah huruf kapital
  data_cp["cap_count"] = data_cp["name"].apply(lambda x: sum(1 for c in x if c.isupper()))

  # Jumlah huruf non kapital
  data_cp["noncap_count"] = data_cp["name"].apply(lambda x: sum(1 for c in x if c.islower()))

  # Jumlah digit
  data_cp["digit_count"] = data_cp["name"].apply(lambda x: sum(1 for c in x if c.isdigit()))

  # Remove punctuation
  data_cp["name"] = data_cp["name"].apply(lambda x: re.sub(r"[^a-zA-Z0-9]", " ", x))

  # Word count
  data_cp["word_count"] = data_cp["name"].apply(lambda x: len(word_tokenize(x)))

  def normalization(text):
    text = text.lower()
    for pattern, replacement in pattern_norm.items():
      text = re.sub(pattern, replacement, text)
    return text

  # Normalization
  if use_normalization:
    data_cp["name"] = data_cp["name"].apply(normalization)

  return data_cp
```

```python
electronics_names = electronics_data_converted[["name"]]
beauty_names = beauty_data_converted[["name"]]

# Normalisasi
electronics_pattern = {
    r"\b[a-z]\d+[a-z]*\b": " SERIES ", #SERIES: A50 A200s B300C
    r"\b[a-z]*\s?\d+[a-z]*\b": " SERIES ", #SERIES: Note 17 iPhone 12
    #r"\b\d+\.?\d?[a-z]*\b": " SERIES ", #SERIES: 5G 7.5inch
    r"\b(pro|promax|max|plus|premium)\b": " SERIES ", #SERIES: pro promax
    r"(samsung)|(vivo)|(xiaomi)|(redmi)|(oppo)|(iphone)|(poco)|(mipow)|(yesplus)|(infinix)": " BRAND ",
}

beauty_pattern = {
     r"\b[a-z]*\d+[a-z]*\b": " SERIES ", #SERIES: A50 A200s B300C
     r"\b(pro|promax|max|plus|premium)\b": " SERIES ", #SERIES: pro promax
     #r"\b\d+\.?\d?[a-z]*\b": " SERIES ", #SERIES: 5G 7.5inch
     r"\b((sweety)|(aftermyskin)|(afterskin)|(bbd)|(nuface)|(nu)|(gloow)|(be)|(morris)|(timephoria)|(implora)|(facetology)|(medikon)|(endzi)|(implora)|(scora)|(close up)|(nutrishe)|(laushine)|(glad2glow)|(maroli)|(glowsophy)|(garnier)|(purbasari)|(scarlett))\b": " BRAND ",

}

electronics_names_extracted = text_extraction(electronics_names, electronics_pattern, True)
beauty_names_extracted = text_extraction(beauty_names, beauty_pattern, True)
```

## Dinov2 (image, 384)

```python
electronics_names_images = pd.concat([electronics_names_extracted["name"], electronics_data_converted["image"]], axis=1)
beauty_names_images = pd.concat([beauty_names_extracted["name"], beauty_data_converted["image"]], axis=1)
```

```python
import torch
from torchvision import transforms
from PIL import Image
import requests
from io import BytesIO

def load_image(transform, img: str) -> torch.Tensor:
  response = requests.get(img)
  try:
    img = Image.open(BytesIO(response.content)).convert("RGB")
  except Exception:
    raise Exception("Gagal buka gambar")
  img = transform(img).unsqueeze(0)
  return img

def image_to_embedding(model, device, data: pd.Series):
  if isinstance(data, pd.DataFrame):
    data = data["image"]

  image_embeddings = []

  with torch.no_grad():
    for link in data.values:
      embeddings = model(load_image(trans, link).to(device))

      embedding = embeddings[0].cpu().numpy()

      image_embeddings.append(embedding)

  return np.array(image_embeddings)

dino = torch.hub.load("facebookresearch/dinov2", "dinov2_vits14")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
dino.to(device)
dino.eval()

trans = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5],
                         std=[0.5, 0.5, 0.5])
])

electronics_image_embeddings = image_to_embedding(dino, device, electronics_names_images["image"])
beauty_image_embeddings = image_to_embedding(dino, device, beauty_names_images["image"])

electronics_img_emb_df = pd.DataFrame(electronics_image_embeddings, columns=[f"emb_img_{i}" for i in range(electronics_image_embeddings.shape[1])])
beauty_img_emb_df = pd.DataFrame(beauty_image_embeddings, columns=[f"emb_img_{i}" for i in range(beauty_image_embeddings.shape[1])])

electronics_concat_img_emb = pd.concat([electronics_concat_emb, electronics_img_emb_df], axis=1)
beauty_concat_img_emb = pd.concat([beauty_concat_emb, beauty_img_emb_df], axis=1)
```
